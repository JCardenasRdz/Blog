```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Function
def T2decay(decay_rate, xdata):
    return np.exp(-decay_rate * xdata)

# variables
x_data = np.linspace(0.005,.20,50)
decay_times = np.linspace(.010,0.080,10)
mu = 0 ; sigma = 0.01

# data frame
df = pd.DataFrame()

# allocate
for  idx, time in enumerate(decay_times):
    df['Sample_'+(str(idx+1))]=T2decay(1/time,x_data) + np.random.normal(mu , sigma, 50)

df.set_index(x_data)
df.index.name='x_data';

df.plot(); plt.show()

df.to_csv('T2_data.csv')
```
